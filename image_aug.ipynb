```python
import numpy as np 
from PIL import Image
import pandas as pd 
import imgaug as ia
import imgaug.augmenters as iaa
from glob import glob
import matplotlib.pyplot as plt
```

```python
seq = iaa.Sequential([
    iaa.Fliplr(0.5),
    iaa.Affine(
               translate_percent={"x": (-0.2, 0.2), "y": (-0.2, 0.2)})
], random_order=True)
```

```python
images_path =  glob("../../data/ImageAug/standardImage/*.jpg")
images=np.zeros((len(images_path),256,256),dtype=np.uint8)
for i in range(751):
    img=Image.open(images_path[i]).convert('L')
    img=img.resize([256,256])
    images[i]=np.array(img,dtype=np.int8)
```

```python
for i in range(751,5306):
    index=np.random.randint(0,750)
    images_aug = seq(images=images[index])
    image_aug=Image.fromarray(images_aug).convert('RGB')
    image_aug.save('../../data/ImageAug/standardImage/'+str(i)+'.jpg')
```
